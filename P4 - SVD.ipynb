```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
```

```python
raw_data = pd.read_csv("train.csv")
raw_data.head()
```

```python
pivot_ii= raw_data.pivot(index="movieID", columns= "userID", values="rating")
pivot_ii.head()
```

```python
all_mean = pivot_ii.mean().mean()
```

```python
pivot_ii.shape
```

```python
scaled_df = pivot_ii - pivot_ii.mean(axis=1).values.reshape(-1, 1)
scaled_df.fillna(0, inplace=True)
scaled_df.head()
```

```python
test = scaled_df[:500]
```

```python
U, S, VT = np.linalg.svd(test, full_matrices=True)
print("U shape:", U.shape) # movies to concepts matrix
print("S shape:", S.shape) # strength vector
print("VT shape:", VT.shape) # concepts to users matrix
```

```python
S_square = S * S
S_square_sum = S_square.sum()
for i in range(1, len(S)):
    res = S_square[:i].sum()/S_square_sum
    if res > 0.90:
        print(i, res)
```

```python
k = 220
U = U[:, :k]
S = S[:k]
VT = VT[:k, :]
```

```python
print("U shape:", U.shape) # movies to concepts matrix
print("S shape:", S.shape) # strength vector
print("VT shape:", VT.shape) # concepts to users matrix
```

```python
U.dot(np.diag(S).dot(VT))
```

```python
len(U.dot(np.diag(S).dot(VT)))
```

```python
results = U.dot(np.diag(S).dot(VT))
results = pd.DataFrame(results,columns=test.columns,index = test.index)
results
```

```python
test_raw = pivot_ii[:500]
```

```python
results2 = results + test_raw.mean(axis=1).values.reshape(-1, 1)
results2
```

```python
for i in results2.columns:
    results2[i] = np.where(results2[i] < 1, 1, results2[i])
    results2[i] = np.where(results2[i] > 5, 5, results2[i])
```

```python
results2.isnull().sum().sum()
```

```python
idx = []
users = []
for user in test_raw.columns:
    for movie in test_raw.index:
        if np.isnan(test_raw.loc[movie,user]) == False:
            idx.append(movie)
            users.append(user)
```

```python
real_ratings = []
for user in test_raw.columns:
    for movie in test_raw.index:
        if np.isnan(test_raw.loc[movie,user]) == False:
            real_ratings.append(test_raw.loc[movie,user])
```

```python
pred_ratings = []
for i in range(len(idx)):
    pred_ratings.append(results2.loc[idx[i],users[i]])
```

```python
len(real_ratings)
```

```python
len(pred_ratings)
```

```python
mean_squared_error(real_ratings,pred_ratings)
```
